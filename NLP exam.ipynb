```python
def gpt4_query(prompt, 
               model="gpt-4",
               system_message="You are a helpful assistant.",
               temperature=0,
               max_tokens=100):
    response = openai.ChatCompletion.create(
        model=model,
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response["choices"][0]["message"]["content"].strip()

# Combine your entire text into one string:
text = """Here is a bag filled with popcorn. 
There is no chocolate in the bag.
Yet, the label on the bag says 'chocolate' and not 'popcorn.' 
Sam finds the bag. She had never seen the bag before.
She cannot see what is inside the bag. She reads the label.
"""

# Ask a single question at the end, and request a short answer:
final_prompt = f"""{text}

In a few words: What does Sam think is in the bag?
"""

final_answer = gpt4_query(final_prompt, max_tokens=10)
print("Final answer:", final_answer)

```
